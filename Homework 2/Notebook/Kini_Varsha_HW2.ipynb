1.(c) Fit a simple linear regression model to predict the response for each predictor. Find if there is statistically significant relationship between the predictors and the response. Create some plots to back up the assertions

```python
# We use statsmodel library instead of scikit-learn because statsmodel has advanced statistical tools for detailed analysis of the results.
# Reference: https://www.statsmodels.org/dev/generated/statsmodels.regression.linear_model.OLS.html
import statsmodels.api as sm
regression_coeff = {}
for i in df.columns[:4]:
    x = df[i]
    y = df['PE']
    X = sm.add_constant(x)
    model = sm.OLS(y,X).fit()
    print("Independent variable: ",i)
    print(model.summary())
    test = model.outlier_test()
    outliers = test[test['bonf(p)'] < 0.05].index.values
    print("Outlier indices:", outliers)
    p = model.params
    regression_coeff[i] = p[1]
    plt.scatter(x,y, color = 'g')
    plt.scatter(x[outliers], y[outliers], color = 'r', label ="Outliers")
    constant = p.const
    slope = p[i]
    plt.plot(x, constant +(slope*x), color = 'k', label ="Regression Line")
    plt.legend()
    plt.xlabel(i)
    plt.ylabel("PE")
    plt.show()
```

The results are described in the model summary. The p-values of the predictors AT, V, AP, RH are equal to 0.000. As a result, all the independent variables are statistically significant for predicting the dependent variable PE.

To back up the assertion, we have individual scatterplots with independent variable on the X-axis and dependent variable on the Y-axis. As it can be seen in the independent variable vs dependent variable scatterplot, the outliers are marked in a different color as compared to the sample points. AT vs PE shows nine outliers whereas only one outlier is seen in the V vs PE scatterplot. AP vs PE and RH vs PE scatterplots do not contain any outliers. The outliers at some indices as specified in the outlier indices list above can be removed from the data for each of the regression task.

Here, (slope*x) + constant gives the prediction for a certain test point x


Showing outliers using Cook's distance

```python
# Reference: https://www.statology.org/how-to-identify-influential-data-points-using-cooks-distance/#:~:text=A%20data%20point%20that%20has,considered%20to%20be%20an%20outlier.
for i in df.columns[:4]:
    x = df[i]
    y = df['PE']
    X = sm.add_constant(x)
    model = sm.OLS(y,X).fit()
    influence = model.get_influence()
    cooks = influence.summary_frame()
    outliers = cooks[cooks['cooks_d'] > 4/9568].index
    print("Outlier indices:", outliers)
    p = model.params
    regression_coeff[i] = p[1]
    plt.scatter(x,y, color = 'g')
    plt.scatter(x[outliers], y[outliers], color = 'r', label ="Outliers")
    constant = p.const
    slope = p[i]
    plt.plot(x, constant +(slope*x), color = 'k', label ="Regression Line")
    plt.legend()
    plt.xlabel(i)
    plt.ylabel("PE")
    plt.show()
```

1.(i) i. Perform k-nearest neighbor regression for this dataset using both normalized and raw features

K-nearest neighbor regression on raw features

```python
# Reference: https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsRegressor.html
from sklearn.neighbors import KNeighborsRegressor
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:, :4], df['PE'], test_size = 0.3)
least_error_train_raw = []
least_error_test_raw = []
for k in range(1,101):
    knn_regressor = KNeighborsRegressor(n_neighbors=k)
    knn_regressor.fit(x_train, y_train)
    pred = knn_regressor.predict(x_test)
    pred_train = knn_regressor.predict(x_train)
    train_error_raw = mean_squared_error(y_train, pred_train)
    test_error_raw = mean_squared_error(y_test, pred)
    least_error_train_raw.append(train_error_raw)
    least_error_test_raw.append(test_error_raw)
k_train_raw = least_error_train_raw.index(min(least_error_train_raw)) + 1
k_test_raw = least_error_test_raw.index(min(least_error_test_raw)) + 1
print("Best Train Fit for raw features is", min(least_error_train_raw),"is at k =", k_train_raw)  
print("Best Test Fit for raw features", min(least_error_test_raw),"is at k =", k_test_raw)
```

K-nearest neighbor regression on normalized features

```python
# Trying to normalize using Min-MaxScaler
# Reference: https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html
from sklearn.preprocessing import MinMaxScaler
norm_train = MinMaxScaler().fit_transform(train)
norm_test = MinMaxScaler().fit_transform(test)
x_norm_train, x_norm_test = norm_train[:,:4], norm_test[:,:4]
y_norm_train, y_norm_test = train.to_numpy()[:,4], test.to_numpy()[:,4]
least_error_train_normalized = []
least_error_test_normalized = []
for k in range(1,101):
    knn_regressor = KNeighborsRegressor(n_neighbors=k)
    knn_regressor.fit(x_norm_train, y_norm_train)
    pred = knn_regressor.predict(x_norm_test)
    pred_train = knn_regressor.predict(x_norm_train)
    train_error = mean_squared_error(y_norm_train, pred_train)
    test_error = mean_squared_error(y_norm_test, pred)
    least_error_train_normalized.append(train_error)
    least_error_test_normalized.append(test_error)
k_train_normalized = least_error_train_normalized.index(min(least_error_train_normalized)) + 1
k_test_normalized = least_error_test_normalized.index(min(least_error_test_normalized)) + 1
print("Best Train Fit for normalized features is", min(least_error_train_normalized),"is at k =", k_train_normalized)  
print("Best Test Fit for normalized features", min(least_error_test_normalized),"is at k =", k_test_normalized)
```
